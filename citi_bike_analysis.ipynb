```python
frames = [jan_2019_df,
          feb_2019_df, 
          mar_2019_df, 
          apr_2019_df,
          may_2019_df, 
          jun_2019_df,
          jul_2019_df,
          aug_2019_df,
          sep_2019_df,
          oct_2019_df,
          nov_2019_df,
          dec_2019_df]
result = pd.concat(frames)
result.head()
```

```python
frames_2020 = [jan_2020_df,
          feb_2020_df,
          mar_2020_df,
          apr_2020_df,
          may_2020_df,
          jun_2020_df,
          jul_2020_df,
          aug_2020_df,
          sep_2020_df,
          oct_2020_df,
          nov_2020_df,
          dec_2020_df]
result_2020 = pd.concat(frames_2020)
result_2020.head()
```

```python
clean_df = result
clean_df['gender'] = clean_df['gender'].apply(str)
clean_df['gender'] = clean_df['gender'].str.replace('0', 'Unknown')
clean_df['gender'] = clean_df['gender'].str.replace('1', 'Male')
clean_df['gender'] = clean_df['gender'].str.replace('2', 'Female')
clean_df.head()
```

```python
clean_df_2020 = result_2020
clean_df_2020['gender'] = clean_df_2020['gender'].apply(str)
clean_df_2020['gender'] = clean_df_2020['gender'].str.replace('0', 'Unknown')
clean_df_2020['gender'] = clean_df_2020['gender'].str.replace('1', 'Male')
clean_df_2020['gender'] = clean_df_2020['gender'].str.replace('2', 'Female')
clean_df_2020.head()
```

```python
clean_df['starttime'] = clean_df['starttime'].apply(str)
clean_df[['StartTimeDate', 'StartTime']] = clean_df.starttime.str.split(" ",expand=True)
clean_df[['StopTimeDate', 'StopTime']] = clean_df.stoptime.str.split(" ",expand=True)
clean_df[['Year', 'Month', 'Day']] = clean_df.StartTimeDate.str.split("-",expand=True)
clean_df['Month'] = clean_df['Month'].apply(str)
clean_df['Month'] = clean_df['Month'].str.replace('12', 'December')
clean_df['Month'] = clean_df['Month'].str.replace('11', 'November')
clean_df['Month'] = clean_df['Month'].str.replace('10', 'October')
clean_df['Month'] = clean_df['Month'].str.replace('01', 'January')
clean_df['Month'] = clean_df['Month'].str.replace('02', 'February')
clean_df['Month'] = clean_df['Month'].str.replace('03', 'March')
clean_df['Month'] = clean_df['Month'].str.replace('04', 'April')
clean_df['Month'] = clean_df['Month'].str.replace('05', 'May')
clean_df['Month'] = clean_df['Month'].str.replace('06', 'June')
clean_df['Month'] = clean_df['Month'].str.replace('07', 'July')
clean_df['Month'] = clean_df['Month'].str.replace('08', 'August')
clean_df['Month'] = clean_df['Month'].str.replace('09', 'September')
clean_df.head(5)
```

```python
clean_df_2020['starttime'] = clean_df_2020['starttime'].apply(str)
clean_df_2020[['StartTimeDate', 'StartTime']] = clean_df_2020.starttime.str.split(" ",expand=True)
clean_df_2020[['StopTimeDate', 'StopTime']] = clean_df_2020.stoptime.str.split(" ",expand=True)
clean_df_2020[['Year', 'Month', 'Day']] = clean_df_2020.StartTimeDate.str.split("-",expand=True)
clean_df_2020['Month'] = clean_df_2020['Month'].apply(str)
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('12', 'December')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('11', 'November')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('10', 'October')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('01', 'January')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('02', 'February')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('03', 'March')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('04', 'April')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('05', 'May')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('06', 'June')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('07', 'July')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('08', 'August')
clean_df_2020['Month'] = clean_df_2020['Month'].str.replace('09', 'September')
clean_df_2020.head(5)
```

```python
bike_df = clean_df 
bike_df = bike_df[["tripduration", "start station id", "start station name", "start station latitude", "start station longitude", "end station id", "end station latitude", "end station longitude", "bikeid", "usertype", "birth year", "gender", "StartTimeDate", "StartTime", "Year", "Month", "Day"]]
bike_df.head(5)
```
